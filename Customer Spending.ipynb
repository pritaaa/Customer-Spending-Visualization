# Distribution of Spending by Nation

```python
df.groupby('country')['spending'].sum().reset_index().sort_values(by='spending', ascending=False)
```

# Distribution of Gender by Spending​

```python
df['gender'].value_counts()
```

```python
df.groupby('gender')['spending'].sum().reset_index()
```

# Comparison Of Spending And Income​

```python
df.groupby('age')[['spending', 'income']].sum().reset_index().sort_values(by='age')
```

```python
comp = df.groupby('age')[['income', 'spending']].sum().reset_index()
comp_sort = comp.sort_values(by='income', ascending=False)
print(comp_sort)
```

# Correlation Between Spending And Purchase Frequency

```python
pr=df['purchase_frequency'].value_counts()
print(pr)
```

```python
cor = df[['purchase_frequency', 'spending']].groupby('purchase_frequency').sum().reset_index()
cor = cor.sort_values(by='purchase_frequency', ascending=False)
print(cor)
```
